```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with your model metrics
data = {
    'Model': ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2'],
    'Train': [0.4767, 0.8983, 0.9717, 0.8983, 0.9717],
    'Accuracy': [0.5075, 0.895, 0.963, 0.721, 0.965],
    'Recall': [0.5048, 0.8867, 0.9407, 0.4163, 0.913],
    'Precision': [0.5327, 0.9045, 0.9061, 1.0, 0.8873]
}

df = pd.DataFrame(data)

# Plot the metrics as lines
plt.figure(figsize=(10, 6))
for metric in ['Accuracy', 'Recall', 'Precision']:
    plt.plot(df['Model'], df[metric], marker='o', label=metric)

plt.title('Model Performance (Line Plot)')
plt.xlabel('Model')
plt.ylabel('Score')
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
